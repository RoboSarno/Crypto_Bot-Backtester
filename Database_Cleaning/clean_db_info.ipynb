## Helper

```python
def replace_db(df, table_name): 
    """_summary_
        replace a db with sqlalchemy
    Args:
        df (pd.DataFrame): df to push to database
        table_name (str): table name in database
    """
    myeng = create_engine(url)

    dbConnection = myeng.connect()
    df.to_sql(con=myeng, name=table_name,  if_exists='replace', index=False)
    dbConnection.close()
```
